**Traccia:** Trasforma un set di dati strutturati (come un dizionario Python) in una descrizione testuale completa.

**Obiettivo:** Convertire dati strutturati in una descrizione leggibile.

```python
%pip install python-dotenv openai
```

```python
import os
import json
from dotenv import load_dotenv
from openai import OpenAI


load_dotenv('.env') 


client = OpenAI()                                                                                                                                       


data = {"name": "Giovanni", "age": 30, "city": "Roma", "profession": "Ingegnere"}


dati_in_formato_testo = json.dumps(data, ensure_ascii=False)


completion = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {
            "role": "system",
            "content": (
                "Sei un copywriter esperto. Il tuo compito è prendere un set di dati strutturati "
                "(in formato JSON/Dizionario) forniti dall'utente e trasformarli in una descrizione "
                "testuale completa, fluida, naturale e discorsiva della persona. "
                "Non limitarti a fare un elenco, ma scrivi un breve paragrafo armonioso. "
                "Fornisci SOLO la descrizione finale, senza introduzioni o commenti."
            )
        },
        {
            "role": "user",
            "content": dati_in_formato_testo 
        }
    ],
)


print("\n--- DESCRIZIONE GENERATA DAI DATI ---")
print(completion.choices[0].message.content)
```
