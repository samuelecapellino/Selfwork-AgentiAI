**Traccia:** Crea un prompt che riassuma un articolo di notizie in un singolo paragrafo di al massimo 255 caratteri.

**Obiettivo:** Prendere un testo lungo e condensarlo nei punti essenziali.

```python
%pip install python-dotenv openai
```

```python
import os
from dotenv import load_dotenv
from openai import OpenAI


load_dotenv('.env') 


client = OpenAI()                                                                                                                                       


input_text = input("Inserisci l'articolo di notizie da riassumere: ")

completion = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {
            "role": "system",
            "content": (
                "Sei un giornalista esperto nella sintesi di notizie per i social media. "
                "Il tuo compito è riassumere l'articolo fornito dall'utente in un singolo paragrafo. "
                "REGOLA FONDAMENTALE E RIGIDA: Il riassunto deve essere lungo al massimo 255 caratteri (spazi inclusi). "
                "Sii conciso, elimina i dettagli superflui e vai dritto al punto principale. "
                "Fornisci SOLO il testo del riassunto, senza contatori, virgolette o commenti."
            )
        },
        {
            "role": "user",
            "content": input_text
        }
    ],
)


riassunto = completion.choices[0].message.content
print("\n--- RIASSUNTO (Max 255 caratteri) ---")
print(riassunto)
print(f"\n[Verifica lunghezza: {len(riassunto)} caratteri]")  #Studiando anche con la AI, mi aveva suggerito di mettere questa riga di codice, cosi che conta automaticamente quanti caratteri ha generato l'AI. Così puoi controllare subito se ha rispettato il vincolo senza dover contare a mano!
```
