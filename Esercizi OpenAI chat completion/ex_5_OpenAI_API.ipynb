**Traccia:** Scrivi un prompt che generi domande di comprensione su un breve testo.

**Obiettivo:** Sviluppare un prompt per creare domande educative basate su un contenuto.

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

# 1. Carica le variabili d'ambiente dal file .env esistente
load_dotenv('.env') 

# 2. Inizializza il client di OpenAI
client = OpenAI()                                                                                                                                       

# 3. Chiede all'utente di inserire il testo su cui generare le domande
input_text = input("Inserisci il testo per generare le domande di comprensione: ")

# 4. Invia la richiesta a gpt-4o-mini con il prompt educativo
completion = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {
            "role": "system",
            "content": (
                "Sei un insegnante esperto nella creazione di test di verifica. "
                "Il tuo compito è leggere il testo fornito dall'utente e generare esattamente "
                "3 domande a risposta multipla per verificare la comprensione del testo. "
                "Per ogni domanda inserisci:\n"
                "- Il testo della domanda\n"
                "- 4 opzioni di risposta (A, B, C, D)\n"
                "- L'indicazione della risposta corretta con una brevissima spiegazione.\n"
                "Rendi le domande stimolanti, basate su fatti realmente presenti nel testo, "
                "e mantieni un tono chiaro e professionale."
            )
        },
        {
            "role": "user",
            "content": input_text
        }
    ],
)

# 5. Stampa il risultato finale
print("\n--- DOMANDE DI COMPRENSIONE GENERATE ---")
print(completion.choices[0].message.content)
```
