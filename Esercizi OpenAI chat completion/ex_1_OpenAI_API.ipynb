**Traccia:** Scrivi un prompt che traduca un testo dall'italiano all'inglese e viceversa. Usa il prompt per tradurre una breve descrizione di un prodotto.

**Obiettivo:** Sviluppare un prompt che sia adatto alla traduzione fluida tra italiano e inglese.

```python
%pip install python-dotenv openai
```

```python
import os
from dotenv import load_dotenv
from openai import OpenAI


load_dotenv('.env') 
client = OpenAI()                                                                                                                                       

input_text = input("Inserisci la descrizione del prodotto da tradurre: ")
completion = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {
            "role": "system",
            "content": (
                "Sei un traduttore professionista bilingue (italiano-inglese). "
                "Il tuo compito è rilevare la lingua del testo fornito dall'utente. "
                "Se il testo è in italiano, traducilo in un inglese naturale, commerciale e fluido. "
                "Se il testo è in inglese, traducilo in un italiano naturale, commerciale e fluido. "
                "Fornisci SOLO la traduzione finale, senza preamboli, spiegazioni o commenti."
            )
        },
        {
            "role": "user",
            "content": input_text
        }
    ],
)


print("\n--- RISULTATO DELLA TRADUZIONE ---")
print(completion.choices[0].message.content)
```
